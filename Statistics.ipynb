## Statistics for Data Science and Python

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats
```

```python
boston_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
boston_df=pd.read_csv(boston_url)
```

```python
MEDV = sns.boxplot(y = 'MEDV', data = boston_df)
MEDV.set(ylabel = 'thousand dollars', title = "Median value of owner-occupied homes in Boston", ylim = (0,55))
```

```python
copy = boston_df.copy()
new_list = copy['CHAS'].replace(0.0,"tract doesn't bound river")
new_list = new_list.replace(1.0, 'tract bound river')
copy['CHAS'] = new_list
CHAS = sns.countplot(x = 'CHAS', data = copy)
CHAS.set(xlabel = 'houses', title = 'Number of houses that bounds Charles River')
```

```python
boston_df['age_group'] = 'NaN'

boston_df.loc[(boston_df['AGE'] <= 35),'age_group'] = '35 and under'
boston_df.loc[(boston_df['AGE']>35)&(boston_df['AGE']<70),'age_group'] = 'between 35 and 70'
boston_df.loc[(boston_df['AGE']>70),'age_group'] = '70 and above'
plot = sns.boxplot(y = 'MEDV', x = 'age_group', data = boston_df)
plot.set(ylabel = 'thousand of dollars', xlabel = 'age groups',
         title = 'Median Value of Owner-occupied Homes in Boston within Different Age Groups')
```

```python
scat_plot = sns.scatterplot(x= 'NOX', y='INDUS', data = boston_df)
scat_plot.set(title = 'Relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town',
             xlabel = 'Nitric Oxide Concentration', ylabel = 'non-retail business acres per town')
```

```python
plt.hist(x = 'PTRATIO', data = boston_df)
plt.title('Histogram of Pupil to Teacher Ratio in Boston')
plt.xlabel('pupil to teacher ratio')
plt.ylabel('count')
```
